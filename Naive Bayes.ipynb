```python
import pandas as pd
import numpy as np
car_df = pd.read_csv(r"C:\Users\drobey\Documents\Coding\Machine Learning - Materials\K579ML\Week 6\car.csv")
car_df.head()
#car_df.dtypes
car_df["buying"] = car_df["buying"].astype('category')
car_df["maint"] = car_df["maint"].astype('category')
car_df["lug_boot"] = car_df["lug_boot"].astype('category')
car_df["safety"] = car_df["safety"].astype('category')
car_df["persons"] = car_df["persons"].astype('category')
car_df["doors"] = car_df["doors"].astype('category')





car_df.dtypes
#car_df.apply(car_df.astype('category'),axis=1) - this doesnt work. Gotta be a way to automate the below
car_df["buying"] = car_df["buying"].cat.codes
car_df["maint"] = car_df["maint"].cat.codes
car_df["lug_boot"] = car_df["lug_boot"].cat.codes
car_df["safety"] = car_df["safety"].cat.codes
car_df["persons"]s = car_df["persons"].cat.codes
car_df["doors"] = car_df["doors"].cat.codes
car_df.head()

```

```python
from sklearn import preprocessing
from sklearn.utils import shuffle
car_numbers = car_df.iloc[:,0:6]
car_numbers.head()
x = car_numbers.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
Car_df=pd.DataFrame(x_scaled, columns=car_numbers.columns)
car_norm = Car_df.join(car_df[["evaluation"]])
car_norm
#car_norm = shuffle(car_norm)
#car_norm.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(car_norm, test_size = .3)
train.head()
train_points = train.iloc[:,0:6]
train_labels = train.iloc[:,6]
test_points = test.iloc[:,0:6]
test_labels = test.iloc[:,6]
train_points.head()
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(train_points,train_labels)
labels = clf.predict(test_points)
clf.score(test_points,test_labels)
#Not so good lol
from sklearn.utils.multiclass import unique_labels
classes = unique_labels(test_labels)
#categories = car_norm.evaluation.unique().tolist() - the function above aligns the labels to match the confusion matrix

print(classes)
```

```python
from sklearn.metrics import confusion_matrix
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
mat = confusion_matrix(test_labels, labels)

sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=classes, yticklabels=classes)
plt.xlabel('true label')
plt.ylabel('predicted label')
#confusion matrix plotted with heatmap
#Note very good
```
