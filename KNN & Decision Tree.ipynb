```python
from sklearn import preprocessing
iris_numbers = Iris_df.iloc[:,0:4]
iris_numbers.head()
```

```python
x = iris_numbers.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
iris_df=pd.DataFrame(x_scaled, columns=iris_numbers.columns)
iris_df.head()
```

```python
Iris_norm = iris_df.join(Iris_df[["class"]])
Iris_norm = shuffle(Iris_norm)
#Bless up, this is how you get the normalized data in with all the column headings and stuff, even if there's a non
#number column. Way harder than mutate lol
Iris_norm.head()
```

```python
#Import the KNN object
from sklearn.neighbors import KNeighborsClassifier
#create the object
classifier = KNeighborsClassifier(n_neighbors = 3)
#Redo the test/train split with normalized data
train, test = train_test_split(Iris_norm, test_size = .3)
train.head()
```

```python
#Now we train the model
training_points = train.iloc[:,0:4]
training_labels = train.iloc[:,4]
classifier.fit(training_points,training_labels)
test.head()
```

```python
test_points = test.iloc[:,0:4]
test_labels = test.iloc[:,4]
classifier.score(test_points,test_labels)
#How accurate the model is on the test data
```

```python
import matplotlib.pyplot as plt
list =[]
for k in range(1,100):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(training_points,training_labels)
    x = classifier.score(test_points,test_labels)
    list.append(x)
print(max(list))
print(list)
X = [i for i in range(99)]
plt.plot(X,list)
plt.show()
#This is the max score for ks 1-100. Gotta be careful of overfitting, don't want to go too high
```

```python
from sklearn.metrics import confusion_matrix
guesses = classifier.predict(test_points)
print(guesses)
```

```python
confusion_matrix(test_labels,guesses)
```
